```python
def binning_by_range(data, bins):
    min_val = min(data)
    max_val = max(data)
    step = (max_val - min_val) / bins
    labels = ['Short', 'Medium', 'Tall']
    result = []

    for value in data:
        bin_index = int((value - min_val) / step)
        if bin_index == bins:
            bin_index -= 1
        result.append(labels[bin_index])

    return result

plant_heights = [10, 12, 15, 22, 25, 27, 32, 34, 38, 42]
print(binning_by_range(plant_heights, 3))
```

```python
humidity = [40, 42, 45, 50, 55, 60, 62]

def min_max(data):
    low = min(data)
    high = max(data)
    norm = []
    for val in data:
        scaled = round((val - low) / (high - low), 2)
        norm.append(scaled)
    return norm

print(min_max(humidity))
```

```python
method_A_scores = [72, 75, 78, 76, 74]
method_B_scores = [68, 65, 66, 69, 67]

def mean(data):
    return sum(data) / len(data)

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / (len(data) - 1)

def t_test_independent(sample1, sample2):
    n1, n2 = len(sample1), len(sample2)
    mean1, mean2 = mean(sample1), mean(sample2)
    var1, var2 = variance(sample1), variance(sample2)

    t = (mean1 - mean2) / (((var1/n1) + (var2/n2)) ** 0.5)
    return t

t_stat = t_test_independent(method_A_scores, method_B_scores)

df = len(method_A_scores) + len(method_B_scores) - 2

print("T-test statistic:", round(t_stat, 3))
print("Degrees of Freedom:", df)

if abs(t_stat) > 2.101:
    print("Result: Significant difference between the two teaching methods.")
else:
    print("Result: No significant difference.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = [
    {"Person": "P1", "Gender": "Boy", "FavoriteColor": "Red"},
    {"Person": "P2", "Gender": "Boy", "FavoriteColor": "Blue"},
    {"Person": "P3", "Gender": "Girl", "FavoriteColor": "Red"},
    {"Person": "P4", "Gender": "Girl", "FavoriteColor": "Blue"},
    {"Person": "P5", "Gender": "Girl", "FavoriteColor": "Blue"},
]

df = pd.DataFrame(data)

print("Dataset:")
print(df.head())

observed = pd.crosstab(df['Gender'], df['FavoriteColor'])
print("\nObserved Frequency Table:")
print(observed)

chi2, p, dof, expected = chi2_contingency(observed)

expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
print("\nExpected Frequency Table:")
print(expected_df.round(2))

print(f"\nChi-square Statistic: {round(chi2, 2)}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {round(p, 4)}")

if p < 0.05:
    print("Conclusion: There is a significant relationship between Gender and FavoriteColor.")
else:
    print("Conclusion: There is no significant relationship between Gender and FavoriteColor.")
```

```python
actual = ['Yes', 'No', 'Yes', 'No', 'Yes']
predicted = ['Yes', 'Yes', 'Yes', 'No', 'No']

def build_confusion_matrix(actual, predicted):
    labels = ['Yes', 'No']
    matrix = [[0, 0], [0, 0]]

    for a, p in zip(actual, predicted):
        i = labels.index(a)
        j = labels.index(p)
        matrix[i][j] += 1
    return matrix

cm = build_confusion_matrix(actual, predicted)
print("Confusion Matrix:")
print("TP:", cm[0][0], "FP:", cm[1][0])
print("FN:", cm[0][1], "TN:", cm[1][1])
```
