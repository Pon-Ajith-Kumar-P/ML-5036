```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Exercise 1
#Loading the dataset
data_cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE']
df = pd.read_csv("housing.csv", header=None, delimiter=r"\s+", names=data_cols)
df
```

```python
print("Features of the dataset")
for i in df.columns[:-1]:
  print(i)
print("\nTarget of the dataset")
print(df.columns[-1])
```

```python
#Shape of the dataset
print("Shape = ",df.shape)
```

```python
#Exercise 2

#Checking for the null values
df.isnull().sum()
```

```python
#Information about the dataset
df.info()
```

```python
#Plotting histogram
df.hist(bins=30, figsize=(15,12))
plt.show()
```

```python
#Plotting pairplots
sns.pairplot(df[['CRIM','RM','LSTAT','PRICE']])
plt.show()
```

```python
#Correlation matrix
corr = df.corr()
corr
```

```python
#Heatmap visualization
plt.figure(figsize=(12,8))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
#Exercise 3

#RM is highly correlated with PRICE
X = df['RM']
y = df['PRICE']
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#Scalarization of the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.values.reshape(-1,1))
X_test = scaler.transform(X_test.values.reshape(-1,1))
```

```python
#Fit the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Intercept
print("Intercept = ",model.intercept_)
print("Slope = ",model.coef_)
```

```python
#Prediction
y_pred = model.predict(X_test)
```

```python
#Plot
plt.figure(figsize=(8,6))
plt.scatter(X_test,y_test,color='green')
plt.plot(X_test,y_pred,color='blue')
plt.xlabel('RM')
plt.ylabel('PRICE')
plt.title('Linear Regression')
plt.show()
```

```python
#Exercise 5
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, classification_report
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R2 Score:", r2)
```

```python
#Exercise 4 (Multi Linear Regression)

X = df.drop('PRICE', axis=1)
y = df['PRICE']
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#Scalarization of the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
#Fit the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_s,y_train)
```

```python
print("Intercept = ",model.intercept_)
print("Slope = ",model.coef_)
```

```python
y_pred = model.predict(X_test_s)
```

```python
#Exercise 5
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, classification_report
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R2 Score:", r2)
```
